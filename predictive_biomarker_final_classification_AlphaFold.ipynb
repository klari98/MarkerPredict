```python
#data preparation
def prepare_data_feature_names(training=training, unknown=unknown, to_drop= to_drop):
    """
    This function prepares the training and unknown data for the machine learning.
    Drops the previously selected columns using the to_drop list without renaming the remaining columns.
    Splits the training data to input parameters and labels.
    Lists the column of input parameters.

    Args:
        to_drop: The list of column names which are going to be dropped from the training and unknown dataframe. 
        training: The dataframe of input parameters which are intended to use to the training of the machine learning model.
        unknown: The dataframe of input parameters which are intended to use to the testing of the machine learning model.

    Returns:
        U: The input data from the unknown dataframe.
        X: The input data from the training dataframe.
        y: The labels for the training dataframe.
        feature_names: The list of column names in the X dataframe.

    """
    training_deleted = training.drop(columns= to_drop)
    U = unknown.drop(columns= to_drop)
    X = training_deleted.iloc[:,:-1]
    y = training_deleted.iloc[:, -1]
    feature_names = list(X.columns)
    return U, X, y, feature_names
```

```python
#teaching - ROC function
def teach_print_train_ROC_CM(classifier, X, y):
    """
    This function teaches and tests the selected machine learning models while drawing the ROC curve and confusion matrix.
    Fits StandardScaler on the training data and transforms the training data.
    Makes predictions on the training dataframe.
    Draws the ROC curve and confusion matrix.

    Args:
        X: The input data which will be used for training and testing the model.
        y: The labels for the X dataframe.
        classifier: The choosen classification algorhythm.

        
    Returns:
        classifier_pred_train: Predictions on the training dataset.

    """
    #scale fitting
    scale = StandardScaler()
    scale.fit(X)
    X = scale.transform(X)
        
    #teaching
    classifier.fit(X, y)

    classifier_pred_train = classifier.predict(X)
    
    print(classifier)
    
    #train ROC curve
    ax1 = plt.gca()
    classifier_disp = RocCurveDisplay.from_estimator(classifier, X, y, ax=ax1, alpha=0.8)
    
    #train CM
    cm = confusion_matrix(y, classifier_pred_train)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()

    plt.show()
    
    return classifier_pred_train
```

```python
#teaching the model to predict on the unknown data - ROC function
def teach_print_unknown(best_classifier, best_classifier_name, X, y, U):
    """
    This function can teach the selected machine learning model on the training and predict on the unknown dataset.
    Currently not used.
    Fits StandardScaler on the training data and transforms the training and unknown data.
    Makes predictions on the unknown dataframe.
    Exports the data to an Excel file.

    Args:
        best_classifier: The best classifier choosed based on previous predictions.
        best_classifier_name: The name of the best classifier.
        X: The input data which will be used for training the model.
        y: The labels for the X dataframe.
        U: The input data which will be used for predicting on the model.

        
    Returns:
        None

    """
    #fitting the scale
    scale = StandardScaler()
    scale.fit(X)
    X = scale.transform(X)
    U = scale.transform(U)
        
    #teaching
    best_classifier.fit(X, y)

    pred_unknown = best_classifier.predict(U)
    pred_unknown_proba = best_classifier.predict_proba(U)
    
    #combining results
    unknown_predicted = unknown.copy()
    unknown_predicted[f"prediction {best_classifier_name}"] = pred_unknown
    unknown_predicted[f"prediction probability {best_classifier_name} 0"] = pred_unknown_proba[:, 0]
    unknown_predicted[f"prediction probability {best_classifier_name} 1"] = pred_unknown_proba[:, 1]
    print(unknown_predicted.head())
    
    #exporting results
    unknown_predicted.to_excel(f'predictions_AlphaFold_{best_classifier_name}.xlsx')
```

```python
#rfc decision tree
def rfc_decision_tree(feature_names, rfc=rfc):
    """
    This function draws the final decision tree generated by the previously trained Random Forest classifier.
    Uses the list of feature names to label the leafs.

    Args:
        feature_names: The list of feature names in the training dataset.
        rfc: The previously trained Random Forest classifier.

        
    Returns:
        None

    """
    tree_rfc = rfc.estimators_[1]
    plt.figure(figsize=(12,12))
    tree.plot_tree(tree_rfc,
                   feature_names = feature_names, 
                   class_names= ("no", "yes"),
                   rounded=True, 
                   filled = True,
                   fontsize= 10)
```

```python
#xgb decision tree + feature importance
def xgb_all(X, y, feature_names):
    """
    This function draws the decision tree and feature importance graph generated by the previously trained XGBoost classifier.
    Uses the list of feature names to label the leafs and the graph..

    Args:
        X: The input data which will be used for drawing the final decision tree and feature importance graph.
        y: The labels for the X dataframe.
        feature_names: The list of feature names in the training dataset.

        
    Returns:
        None

    """

    scale = StandardScaler()
    scale.fit(X)
    X = scale.transform(X)

    d = xgb_package.DMatrix(X, label=y, feature_names=feature_names)
    #teaching
    param = {
            'objective': 'binary:logistic',

        }

    model= xgb_package.train(dtrain=d, params=param)

    rcParams['figure.figsize'] = 15, 7
    xgb_package.plot_tree(model)
    plt.show()


    xgb_package.plot_importance(model)
```
